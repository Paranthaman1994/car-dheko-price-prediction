```python
import pandas as pd
import ast

# List of dictionary strings
dict_strings =  df["new_car_detail"].to_list()


# Convert each dictionary string to a dictionary
data_dicts = [ast.literal_eval(dict_string) for dict_string in dict_strings]

# Convert the list of dictionaries to a DataFrame
df_detail = pd.DataFrame(data_dicts)


```

```python
import pandas as pd
import ast

# Assuming df["new_car_overview"] is a Series of stringified dictionaries

# Function to convert each string to a dictionary
def str_to_dict(row):
    return ast.literal_eval(row)

# Apply the function to the Series
data_dicts = df["new_car_overview"].apply(str_to_dict)

# Now, extract the 'top' data from each dictionary and create a DataFrame
data_list = [{item['key']: item['value'] for item in d['top']} for d in data_dicts]

# Convert the list of dictionaries into a DataFrame
df_overview = pd.DataFrame(data_list)

```

```python
df_overview
```

```python
import pandas as pd
import json

# Example Series data (replace with your actual Series)
data_series = df["new_car_feature"]

def extract_features(data_dict):
    # Initialize lists to hold features for each category
    features_dict = {
        'Comfort': [],
        'Interior': [],
        'Exterior': [],
        'Safety': []
    }
    
    # Extract detailed features
    for category in data_dict.get('data', []):
        sub_heading = category.get('subHeading')
        feature_list = [item['value'] for item in category.get('list', [])]
        if sub_heading in features_dict:
            features_dict[sub_heading].extend(feature_list)
    
    # Combine features into single strings
    combined_features = {key: ', '.join(features_dict[key]) for key in features_dict}
    
    return combined_features

# Process each row in the Series
all_features = []
for json_str in data_series:
    # Convert JSON string to Python dictionary
    try:
        data_dict = json.loads(json_str.replace("'", '"'))
        features = extract_features(data_dict)
        all_features.append(features)
    except json.JSONDecodeError as e:
        print(f"JSON Decode Error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Create DataFrame from the processed features
df_features = pd.DataFrame(all_features)
```
