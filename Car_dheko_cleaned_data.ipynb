```python
import pandas as pd
```

```python
df=pd.read_csv("car_dheko.csv")
```

```python
df
```

```python
df.info()
```

```python

# Threshold for dropping columns (80% missing values)
threshold = 0.8

# Calculate the threshold for the number of missing values
missing_threshold = len(df) * threshold

# Keep a list of original columns
original_columns = df.columns

# Drop columns with more than 80% missing values
df_cleaned = df.dropna(axis=1, thresh=len(df) - missing_threshold)

# Identify deleted columns
deleted_columns = original_columns.difference(df_cleaned.columns)

print("Original columns:", original_columns)
print("Deleted columns:", deleted_columns)
```

```python
df = df_cleaned
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df=df.drop(columns="unnamed:_0")
```

```python
df.isna().sum()
```

```python
df["bt"] = df["bt"].fillna(df["bt"].mode()[0])
```

```python
df['registration_year'] = pd.to_numeric(df['registration_year'], errors='coerce')
df["registration_year"] = df["registration_year"].fillna(df["registration_year"].mean())
```

```python
df["insurance_validity"] = df["insurance_validity"].fillna(df["insurance_validity"].mode()[0])
```

```python
df["seats"] = df["seats"].fillna(df["seats"].mode()[0])
```

```python
df["kms_driven"] = df["kms_driven"].fillna(df["kms_driven"].mode()[0])
```

```python
df["rto"] = df["rto"].fillna(df["rto"].mode()[0])
```

```python
df["ownership"] = df["ownership"].fillna(df["ownership"].mode()[0])
```

```python
df["engine_displacement"] = df["engine_displacement"].fillna(df["engine_displacement"].mode()[0])
```

```python
df["year_of_manufacture"] = df["year_of_manufacture"].fillna(df["year_of_manufacture"].mean())
```

```python
df["mileage"] = df["mileage"].fillna(df["mileage"].mode()[0])
```

```python
df= df.dropna(axis=1, how='any')
```

```python
df.columns
```

```python
columns_to_drop = ["it", "ft", "ownerno", "centralvariantid", "variantname", "trendingtext", "registration_year", "kms_driven", "ownership", "year_of_manufacture","transmission"]
df_dropped = df.drop(columns=columns_to_drop)
```

```python
df=df_dropped
```
